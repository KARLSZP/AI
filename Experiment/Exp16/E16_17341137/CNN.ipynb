```python
train_data = {
  'X_train': data['X_train'],
  'y_train': data['y_train'],
  'X_val': data['X_val'],
  'y_val': data['y_val'],
}
```

# 2 Build the model

As described in the docstring:

A three-layer convolutional network with the following architecture:

> conv - relu - 2x2 max pool - affine - relu - affine - softmax

```python
# For reproducity, set random seed.
np.random.seed(666)
```

```python
model = ThreeLayerConvNet(hidden_dim=256, reg=0.05)
```

# 3  Set up the solver

```python
solver = Solver(model, data,
                update_rule='adam',
                optim_config={'learning_rate': 0.001},
                lr_decay=1.0,
                batch_size=50,
                num_epochs=10,
                print_every=1,
                verbose=True,
               num_train_samples=None)
```

# 4 Training

```python
solver.train()
```

# 5 Benchmarking

## 5.1 Accuracy on traning set

```python
X = train_data['X_train']
y = train_data['y_train']
solver.check_accuracy(X, y)
```

## 5.2 Accuracy on validating set

```python
X = train_data['X_val']
y = train_data['y_val']
solver.check_accuracy(X, y)
```

# 7 Conclusion

In this experiment, I combine those provided components and set up a 3-layer conv NN by hand.

As shown above, the model is defined with the following architecture:

> conv - relu - 2x2 max pool - affine - relu - affine - softmax

A convolution layer to mask each filter with the input image, and then relu layer activate the result on feature maps to generate output of conv layer.

The next layer is named as 'pooling' layer, which use 'pooling' strategy to lower the dimension of the data without losing main/major features. It did work well and effectively reduce the complexity of computation.

Eventully, a fully connected layer(affine layer) is defined to map the result onto a 10-dimension output(in this exp. specially) as a classifier.
